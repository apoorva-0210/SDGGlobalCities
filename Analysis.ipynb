```python
df1 = pd.read_csv('BangkokScoreRAW1.csv')
df2 = pd.read_csv('BeijingScoreRAW1.csv')
df3 = pd.read_csv('IstanbulScoreRAW.csv')
df4 = pd.read_csv('JakartaScoreRAW1.csv')
df5 = pd.read_csv('KarachiScoreRAW.csv')
df6 = pd.read_csv('LagosScoreRAW1.csv')
df7 = pd.read_csv('ManilaScoreRAW1.csv')
df8 = pd.read_csv('MexicoCityScoreRAW1.csv')
df9 = pd.read_csv('AmsterdamScoreRAW1.csv')
df10 = pd.read_csv('SaoPauloScoreRAW1.csv')
df11 = pd.read_csv('SeoulScoreRAW.csv')
#df12 = pd.read_csv('ShanghaiScoreRAW1.csv')
df13 = pd.read_csv('SydneyScoreRAW1.csv')
df14 = pd.read_csv('TokyoScoreRAW.csv')
```

```python
df1['City'] = 'Bangkok'
df2['City'] = 'Beijing'
df3['City'] = 'Istanbul'
df4['City'] = 'Jakarta'
df5['City'] = 'Karachi'
df6['City'] = 'Lagos'
df7['City'] = 'Manila'
df8['City'] = 'MexicoCity'
df9['City'] = 'Amsterdam'
df10['City'] = 'SaoPaulo'
df11['City'] = 'Seoul'
#df12['City'] = 'Shanghai'
df13['City'] = 'Sydney'
df14['City'] = 'Tokyo'
```

```python
import pandas as pd

# Target columns
base_columns = ['Building', 'lon', 'lat', 'City']
score_columns = [
    'Civic_Score', 'Fire_Score', 'Government_Score', 'Hospital_Score',
    'Public_Score', 'School_Score', 'Sports_Score', 'Supermarket_Score',
    'Toilet_Score', 'Transport_Score'
]

distance_columns = ['Civic', 'Fire', 'Government', 'Hospital', 'Public',
                    'School', 'Sports', 'Supermarket', 'Toilet', 'Transport']

# List of dataframes
dfs = [df3, df5, df9, df11, df14]

# Process each dataframe
processed_dfs = []

for df in dfs:
    df = df.copy()

    # Calculate averaged _Score if possible
    for col in distance_columns:
        try:
            df[f'{col}_Score'] = (
                df[f'{col}_1km_Score'] + df[f'{col}_5km_Score'] + df[f'{col}_10km_Score']
            ) / 3
            df[f'{col}_Score'] = df[f'{col}_Score'].round(2)
        except KeyError:
            pass

    # Keep only available score columns
    available_scores = [col for col in score_columns if col in df.columns]

    # Final set of columns to retain
    final_cols = base_columns + available_scores
    df = df[final_cols]

    # Compute TotalScore only with available scores
    df['TotalScore'] = df[available_scores].mean(axis=1)

    processed_dfs.append(df)

# Optional: Combine them if needed
all_processed = pd.concat(processed_dfs, ignore_index=True)
```
